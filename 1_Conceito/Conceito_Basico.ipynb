```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.image as img

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

# CRIANDO BASE DE DADOS

```python
idades = np.random.randint(low=1, high=100, size=500)
idades
```

# RANDOMIZANOD 10% DOS DADOS

```python
labels = []  
for idade in idades:  
    if idade < 30:  
        labels.append(0)  
    else:
        labels.append(1) 

for i in range(0, int(idades.shape[0] * 0.1)):
    r = np.random.randint(0, len(labels) - 1)  
    if labels[r] == 0: 
        labels[r] = 1 
    else:
        labels[r] = 0 
```

```python
plt.scatter(idades, labels, color="red")
plt.show()
```

# PREDIÇÃO USANDO REDE NEURAL

```python
idades.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(idades.reshape(-1, 1), labels)
```

```python
imagem = img.imread('funcao_reta.webp')
plt.imshow(imagem)
plt.axis('off') 
plt.show()
```

```python
# y = m.x + b
m = model.coef_[0]
b = model.intercept_
print(m, b)
```

# CONCEITO DE REGRESSÃO LINEAR

```python
# 0.5 = m.x + b
# 0.5 - b = m.x
#( 0.5 - b) / m = x
limiar_idade = (0.5 - b) / m
print(limiar_idade)

plt.plot(idades, idades * m + b, color='blue', label='Linha de regressão')
plt.plot([limiar_idade, limiar_idade], [0, 0.5], '--', color='green', label='Limiar de idade')
plt.scatter(idades, labels, color='red', label='Dados')

plt.xlabel('Idade')
plt.ylabel('Saída do Modelo')
plt.title('Regressão Linear e Limiar de Idade')
plt.legend()
plt.show()
```

# CONCEITO FUNÇÃO LOGICA

```python
imagem = img.imread('funcao_logitica.png')
plt.imshow(imagem)
plt.axis('off') 
plt.show()
```

```python
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1 / (1 + math.exp(-item)))
    return a

x = np.arange(-10., 10., 0.1)
print(x[:5])
sig = sigmoid(x)
print(sig[:5])
```

```python
plt.plot(x, sig)
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Função Sigmóide')
plt.grid(True)
plt.show()
```

# Classificador Sigmóide

```python
modelo = LogisticRegression()
modelo.fit(idades.reshape(-1, 1), labels)

# coeficientes (m) 
# intercepto (b) 
m = modelo.coef_[0][0]
b = modelo.intercept_[0]

x = np.arange(0, 100, 0.1)
sig = sigmoid(m*x + b)

limiar_idade = -b / m
print(f'X: {x[:5]}')
print(f'SIGMOIDE: {sig[:5]}')
print("Limiar de Idade:", limiar_idade)
```

```python
plt.scatter(idades, labels, color="red", label="Dados")
plt.plot([limiar_idade, limiar_idade], [0, 0.5], '--', color="green", label="Limiar de Decisão")
plt.plot(x, sig, label="Função Sigmoide")

plt.xlabel("Idade")
plt.ylabel("Saída do Modelo")
plt.title("Regressão Logística e Função Sigmoide")
plt.legend()
plt.show()
```

# PERCEPTRON SIMPLES

```python
# colunas são as características (X1, X2, X3) e a última coluna é o rótulo (Y)
data = [[0, 0, 1, 1],
        [1, 0, 1, 1],
        [1, 1, 1, 0],
        [1, 1, 0, 0]]

# iniciando pesos
w = [0, 0, 0, 0]

# Índices das colunas dos dados e do rótulo
X_1 = 0
X_2 = 1
X_3 = 2
Y = 3

# Taxa de aprendizado
learning_rate = 0.1

for iterations in range(100):
    for SAMPLE in range(len(data)):
        # Calcula o valor de z (combinação linear das características ponderadas pelos pesos)
        z = sum(w[i] * data[SAMPLE][i] for i in range(len(w)))
        print("Valor de z:", z)

        # Função de ativação: se z for maior que 0, prevê a classe 1; caso contrário, prevê a classe 0
        pred_y = 1 if z > 0 else 0

        # Atualização dos pesos com base no erro
        for i in range(len(w)):
            w[i] += learning_rate * (data[SAMPLE][Y] - pred_y) * data[SAMPLE][i]

        print("Pesos atualizados:", w)

# Imprime os pesos finais
print("Pesos finais:", w)
```

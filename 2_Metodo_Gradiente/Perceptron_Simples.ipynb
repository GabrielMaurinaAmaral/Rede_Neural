```python
# colunas so as caractersticas (X1, X2, X3) e a ltima coluna  o rtulo (Y)
data = [[0, 0, 1, 1],
        [1, 0, 1, 1],
        [1, 1, 1, 0],
        [1, 1, 0, 0]]

# iniciando pesos
w = [0, 0, 0, 0]

# ndices das colunas dos dados e do rtulo
X_1 = 0
X_2 = 1
X_3 = 2
Y = 3

# Taxa de aprendizado
learning_rate = 0.1

for iterations in range(100):
    for SAMPLE in range(len(data)):
        # Calcula o valor de z (combinao linear das caractersticas ponderadas pelos pesos)
        z = sum(w[i] * data[SAMPLE][i] for i in range(len(w)))
        print("Valor de z:", z)

        # Funo de ativao: se z for maior que 0, prev a classe 1; caso contrrio, prev a classe 0
        pred_y = 1 if z > 0 else 0

        # Atualizao dos pesos com base no erro
        for i in range(len(w)):
            w[i] += learning_rate * (data[SAMPLE][Y] - pred_y) * data[SAMPLE][i]

        print("Pesos atualizados:", w)

# Imprime os pesos finais
print("Pesos finais:", w)
```
